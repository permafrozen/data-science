Es gibt zwei Weg um eine virtuelle Arbeitsumgebung mit Anaconda anzulegen, den schnellen über die **Anaconda Prompt** und den einfachen über die **Anaconda Navigator GUI**.

Ich empfehle den Weg über die Prompt.

#### 1.1.5.1 Anaconda Prompt
- Anaconda Prompt über das Startmenü öffnen
- `conda env list`
- `conda create --name Test_Umgebung_py37 python=3.7 numpy`
  
  
  
- `conda env list`
- `conda activate Test_Umgebung_py37`
  
  Nun sollte in deiner Prompt die letzet Zeile mit **(Test_Umgebung_py37)** beginnen. Somit arbeitest du jetzt in deiner virtuellen Umgebung.
  Jetzt haben wir aber nur das Package numpy installiert, dass ist zwar schön aber noch nicht ausreichend. Wenn du jetzt als Befehle `conda list` eingibst, siehst du,
  dass deine virtEnv nur eine ganz kurze List von Paketten enthält (man hätte gleich beim create Statement weiter Packete mit installieren können,
  indem man sie einfach nach numpy dazu schreibt, z.B. `conda create --name Test_Umgebung_py37 python=3.7 numpy pandas jupyterLab ....`)

  

- `conda install --name Test_Umgebung_py37 pandas jupyterlab`

Erste ***1.Übung (HÜ)*** richte dir eine virtuelle Umgebung mit den Paketten numpy pandas matplotlib seaborn und jupyterlab ein.

Um eine virtuelle Umgebung wieder zu löschen, wähle folgende Anweisung `conda remove --name Test_Umgebung_py37 --all`

Es gibt Package die über den standard Channel nicht installierbar sind, dann gibt es zwei Möglickeiten, entweder man erweitert den Befehl wie folgt `conda install -c conda-forge tensorflow` wobei **conda-forge** der Name des Channels ist über das man das Package laden kann. Der Nachteil dieses Vorgehens ist, dass man bei jeder neuen virtuellen Umgebung wieder diesen langen Befehl eingeben muss. Zweite Möglichkeit ist, man legt zuerst einen neuen Channel an und kann diesen dann immer nutzen. 

 - `conda config --show channels` damit sieht man die bereits vorhanden Channels
 - `conda config --add channels conda-forge`
 - `conda config --show channels`
 
  

   `conda config --get channels` erhält man Informationen, welcher Channel höhere Priorität besitzt.
   
   `conda config --remove channels conda-forge` damit entfernt man den Channel wieder.
    
## 1.2.  JupyterLab
 Ich verwende sehr gern Jupyter Lab anstatt nur Jupyter Notebook, weil es einen größeren Funktionsumfang biete im Vergleich zu Jupyter Notebook. Jupyter Lab ist auch das jüngere der beiden Projekt und ich gehe davon aus, das Jupyter Notebook früher oder später in Jupyter Lab aufgehen wird.

 Die Jupyter Notebooks werden in einem Webbrowser geöffnet, in der Regel in einem localen Host. Besondere Stärke sind die Darstellungsmöglichkeiten der Notebooks und die Kombination aus Code (Code Zellen) und formatierten Text (Markdown Zellen). Mit der Schnittstelle IPython (Interaktives Python) ist der Kernel der für die Codeauswertung verwendet wird, realisiert. Standardmäßig können die Sprachen Julia und R ebenfalls verwendet werden, darauß leitet sich auch der Name Jupyter (**Ju**lia, **Pyt**hon **R**) her.

Die Handhabung der GUI ist selbsterklärend und wird in der Regel mit **learning by doing** sehr schnell erlernt.

Wie ihr seht ist Jupyter Notebook in Zellen organisiert, wobei für uns die zwei Typen Code-Zellen und Markdown-Zellen wesentlich sind. In Code Zellen wird der Python Code eingegeben und mit **SHIFT+ENTER** ausgeführt. Die Markdown zellen sind für den begleitenden Text vorgesehen, wobei man auch LATEX Syntax und HTML verwenden kann.  

Wenn man eine neue Zelle anlegt, ist das eine Code-Zelle, damit man diese Zelle in eine Markdown Zelle umwandelt, nützt man die Tastenkombination **ESC+m**. 

**ESC+y** ermöglicht den umgekehrten Weg von einer Markdown Zelle in eine Code Zelle.

Raw Zellen werden genau so angezeigt wie ihr sie eingebt, siehe folgendes Beispiel (**ESC+r**)

### 1.2.3 Magic Functions

Die erste Funktion %xmode (Plain,Context oder Verbose) ist für einfache Fehleranalysen gedacht. Mit den Parameter steuert man, die Informationsfülle.

```python
%xmode Plain
a=7
b=0
c=a/b
print(c)
```

```python
%xmode Context
a=7
b=0
c=a/b
print(c)
```

```python
%xmode Verbose
a=7
b=0
c=a/b
print(c)
```

Der Debugger wird mit %debug aufgerufen

```python
a=10
b=5
for i in range(10):
    print(a/(b-i))
```
