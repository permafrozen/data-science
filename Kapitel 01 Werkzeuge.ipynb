#### 1.1.5.1 Anaconda Prompt
- Anaconda Prompt ber das Startmen ffnen
- `conda env list`
- `conda create --name Test_Umgebung_py37 python=3.7 numpy`
  
  
  
- `conda env list`
- `conda activate Test_Umgebung_py37`
  
  Nun sollte in deiner Prompt die letzet Zeile mit **(Test_Umgebung_py37)** beginnen. Somit arbeitest du jetzt in deiner virtuellen Umgebung.
  Jetzt haben wir aber nur das Package numpy installiert, dass ist zwar schn aber noch nicht ausreichend. Wenn du jetzt als Befehle `conda list` eingibst, siehst du,
  dass deine virtEnv nur eine ganz kurze List von Paketten enthlt (man htte gleich beim create Statement weiter Packete mit installieren knnen,
  indem man sie einfach nach numpy dazu schreibt, z.B. `conda create --name Test_Umgebung_py37 python=3.7 numpy pandas jupyterLab ....`)

  

- `conda install --name Test_Umgebung_py37 pandas jupyterlab`

Erste ***1.bung (H)*** richte dir eine virtuelle Umgebung mit den Paketten numpy pandas matplotlib seaborn und jupyterlab ein.

Um eine virtuelle Umgebung wieder zu lschen, whle folgende Anweisung `conda remove --name Test_Umgebung_py37 --all`

Es gibt Package die ber den standard Channel nicht installierbar sind, dann gibt es zwei Mglickeiten, entweder man erweitert den Befehl wie folgt `conda install -c conda-forge tensorflow` wobei **conda-forge** der Name des Channels ist ber das man das Package laden kann. Der Nachteil dieses Vorgehens ist, dass man bei jeder neuen virtuellen Umgebung wieder diesen langen Befehl eingeben muss. Zweite Mglichkeit ist, man legt zuerst einen neuen Channel an und kann diesen dann immer nutzen. 

 - `conda config --show channels` damit sieht man die bereits vorhanden Channels
 - `conda config --add channels conda-forge`
 - `conda config --show channels`
 
  

   `conda config --get channels` erhlt man Informationen, welcher Channel hhere Prioritt besitzt.
   
   `conda config --remove channels conda-forge` damit entfernt man den Channel wieder.

### 1.2.3 Magic Functions

Die erste Funktion %xmode (Plain,Context oder Verbose) ist fr einfache Fehleranalysen gedacht. Mit den Parameter steuert man, die Informationsflle.

```python
%xmode Plain
a=7
b=0
c=a/b
print(c)
```

```python
%xmode Context
a=7
b=0
c=a/b
print(c)
```

```python
%xmode Verbose
a=7
b=0
c=a/b
print(c)
```
